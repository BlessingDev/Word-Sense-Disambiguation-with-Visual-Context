# 0410 실험 결과

valid invalid 기준을 명확하게 한 후 이 기준을 4개로 나누어 각각에 대해 example을 주고 분류하도록 하였다. 

````python
import pandas as pd

def calculate_accuracy(df):
    processed_answers = list()
    for index, row in df.iterrows():
        answer = row["answer"]
        
        if "Answer:" in answer:
            processed_answer = answer.split("Answer:")[-1].strip()
        else:
            processed_answer = "incomplete"
        processed_answers.append(processed_answer)

    df["predicted_answer"] = processed_answers

    accuracy = (df["predicted_answer"] == df["label"]).sum() / len(df)
    return accuracy
````

### threshold를 조정하여 성능 높여보기

기본 threshold는 0.5.

정확도 93%

verb(1)를 noun(0)으로 판정한 사례: 6

noun(0)을 verb(1)로 판정한 사례: 1

어느 정도 trade off는 나타나겠지만, threshold를 1쪽으로 더 조정하면 정확도 향상이 있을 수 있음

````python
# probabilities 가져오기
import numpy as np
probabilities = np.load("/workspace/data/dataset_construction_train/inference/pos_tagging_modernbert_large_test_results/test_probabilities.npy")
````

````python
# label 가져오기
import pandas as pd
test_df = pd.read_csv("/workspace/data/dataset_construction_train/pos_tagging_test.csv")
labels = test_df["label"].values
````

````python
decision_threshold = 0.45
predicted_labels = (probabilities[:, 1] > decision_threshold).astype(int)
````

````python
accuracy = (predicted_labels == labels).sum() / len(labels)
print(f"Accuracy: {accuracy}")
````

````python
decision_threshold = 0.3
predicted_labels = (probabilities[:, 1] > decision_threshold).astype(int)
accuracy = (predicted_labels == labels).sum() / len(labels)
print(f"Accuracy: {accuracy}")
````

````python
# label 0에 대한 정확도
negative_accuracy = ((predicted_labels == 0) & (labels == 0)).sum() / (labels == 0).sum()
print(f"Negative Class Accuracy: {negative_accuracy}")
# label 1에 대한 정확도
positive_accuracy = ((predicted_labels == 1) & (labels == 1)).sum() / (labels == 1).sum()
print(f"Positive Class Accuracy: {positive_accuracy}")
````

정확도에 편향성을 확인할 수 있어. class imbalance로 인한 것으로 보이고, 이 문제를 해결하려면... upscaling해봐야 할 것 같네.

## 실험 2 - verb 샘플 증강

verb sample을 증강한 후에 한 실험

````python
# probabilities 가져오기
import numpy as np
probabilities = np.load("/workspace/data/dataset_construction_train/inference/pos_tagging_modernbert_large_augmentedx2_test_results/test_probabilities.npy")
````

````python
# label 가져오기
import pandas as pd
test_df = pd.read_csv("/workspace/data/dataset_construction_train/pos_tagging_test.csv")
labels = test_df["label"].values
````

현재 noun을 verb로 판정한 sample이 더 많다.

따라서 threshold를 높여 noun으로 판정하기 쉽게 만들어 성능을 측정해본다.

````python
decision_threshold = 0.5
predicted_labels = (probabilities[:, 1] > decision_threshold).astype(int)
accuracy = (predicted_labels == labels).sum() / len(labels)
print(f"Accuracy: {accuracy}")
````

결국 93~94의 범위를 벗어나지 못한다.

프롬프트를 개선하여 "verb와 noun이 모두 가능할 때 sense list를 검토하여 문장에서 더 다양하게 사용 가능한 pos로 골라라"라는 내용을 추가하자.

````python
negative_accuracy = ((predicted_labels == 0) & (labels == 0)).sum() / (labels == 0).sum()
print(f"Negative Class Accuracy: {negative_accuracy}")
# label 1에 대한 정확도
positive_accuracy = ((predicted_labels == 1) & (labels == 1)).sum() / (labels == 1).sum()
print(f"Positive Class Accuracy: {positive_accuracy}")
````

````python
tp = ((predicted_labels == 1) & (labels == 1)).sum()
tn = ((predicted_labels == 0) & (labels == 0)).sum()
fn = ((predicted_labels == 0) & (labels == 1)).sum()
fp = ((predicted_labels == 1) & (labels == 0)).sum()

print(f"TP: {tp}, TN: {tn}, FP: {fp}, FN: {fn}")
````

## pos 판별 실험

````python
import pandas as pd

result_df = pd.read_csv("/workspace/data/test_set_process/inference/wsd_100_pos_gemma3-27b.csv")

pos_mapping = {
    "noun": "n",
    "verb": "v",
    "adjective": "a",
    "adverb": "r"
}
for index, row in result_df.iterrows():
    pred_label = row["generated_text"].split("A:")[-1].strip().lower()
    mapped_label = pos_mapping.get(pred_label, "invalid")
    result_df.at[index, "predicted_label"] = mapped_label

acc = (result_df["predicted_label"] == result_df["gold_pos"]).sum() / len(result_df)
print(acc)
````
