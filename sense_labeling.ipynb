## 훈련 셋 베이스 만들기 (통합하기)

```python
labeled_index = labeled_df["word_index"].tolist()
labeled_index = set(labeled_index)
# 삭제한 샘플은 labeled_df에는 들어있지 않음
# entire_df에서 labeled_df에 들어있지 않은 샘플을 찾아서 labled_df에 추가
# labeled_df에는 ambiguous sentence 열이 있는데 entire_df에는 ambiguous sentence 열이 없다.
# 따라서 ambiguous sentence 열을 일종의 label로 사용할 수 있음.
# 비어있으면 취급 안 하는 sample이라는 레이블이 붙은 것으로 간주

import json
# labeled_df에서 가장 마지막으로 label한 샘플까지만 순회를 진행
last_labeled_index = labeled_df["word_index"].max()
entire_df_labeled = entire_df[entire_df["word_index"] <= last_labeled_index]
for index, row in entire_df_labeled.iterrows():
    if row["word_index"] not in labeled_index:
        new_row = pd.DataFrame({
            "word_index": [row["word_index"]],
            "word": [row["word"]],
            "senses": [row["senses"]],
            "gold_image": [row["gold_image"]],
            "gold_sense": [-1],
            "ambiguous_sentence": [""],
        })
        labeled_df = pd.concat([labeled_df, new_row], ignore_index=True)
    else:
        cur_labeled_row = labeled_df[labeled_df["word_index"] == row["word_index"]]
        sense_dict = json.loads(cur_labeled_row["senses"].iloc[0])
        if "noun_senses" in sense_dict:
            sense_dict["gold_pos"] = "noun"
        elif "verb_senses" in sense_dict:
            sense_dict["gold_pos"] = "verb"
        
        entire_sense_dict = json.loads(row["senses"])
        sense_dict.update(entire_sense_dict)
        
        labeled_df.loc[labeled_df["word_index"] == row["word_index"], "senses"] = json.dumps(sense_dict, ensure_ascii=False)

labeled_df
```

### LLM zero-shot 품사 분류 수행을 위한 prompt 템플릿 마련

```python
prompt_template="""
Ambiguous word: {word}
---
Noun senses:
{noun_senses}
---
Verb senses:
{verb_senses}
---
Adjective senses:
{adj_senses}
---
Adverb senses:
{adv_senses}
---
You are an expert linguistic annotator. Image, a specific target word found within the context of that image, and a list of potential senses for that word are given. Each sense is categorized by a Part of Speech (POS): noun, verb, adjective, or adverb.

Your goal is to determine the correct POS by selecting the sense that most directly aligns with the visual context.
---
Decision Logic & Rules
Follow these rules in order of priority to make your decision:

Directness of Meaning: Choose the sense that provides the most direct and specific fit for the visual context. Even if a perfect match does not exist, select the sense that is the closest indirect match.

The "Verb" Action Rule: Assign the Verb POS if the image depicts a person or entity performing the action. A word can be classified as a Verb even if no humans are present, provided the image illustrates the active manifestation or "happening" of that sense.

Crucial Exception: Choose POS of the sense that provides a more direct, varied, or dynamic description of the image than other senses, prioritize the POS.

---
After all your reasoning is finished, provide your final decision as the format 'A: [POS]', where [POS] is your chosen Part of Speech (noun, verb, adjective, or adverb).
"""
```

```python
import pandas as pd
import json

wsd_test_df = pd.read_csv("/workspace/data/set_process/wsd_set_test_labeled.csv")

prompt_list = list()
for index, row in wsd_test_df.iterrows():
    sense_dict = json.loads(row["senses"])
    noun_senses = sense_dict.get("n", [])
    verb_senses = sense_dict.get("v", [])
    adj_senses = sense_dict.get("a", [])
    adv_senses = sense_dict.get("r", [])
    prompt = prompt_template.format(
        word=row["word"],
        noun_senses="\n".join([f"- {sense}" for sense in noun_senses]),
        verb_senses="\n".join([f"- {sense}" for sense in verb_senses]),
        adj_senses="\n".join([f"- {sense}" for sense in adj_senses]),
        adv_senses="\n".join([f"- {sense}" for sense in adv_senses]),
    )
    prompt_list.append(prompt)

wsd_test_df["prompt"] = prompt_list
wsd_test_df.to_csv("/workspace/data/set_process/wsd_set_test_pos_prompt2.csv", index=False)
```

```python
prompt_template="""Text Context: {context}
Ambiguous Word: {word}
---
Sense List:
{sense_list}
---
You are an expert linguistic annotator. A specific target word found within the context of a given text, and a list of potential senses for that word are provided. Your task is to determine the correct sense by selecting the one that most directly aligns with the context and its background event.
---
Decision Logic & Rules
Follow these rules in order of priority to make your decision:

Directness of Meaning: Choose the sense that provides the most direct and specific fit for the visual and linguistic context. Even if a perfect match does not exist, select the sense that is the closest indirect match.

Focus on entire context: Consider the entire context of the image, not just the immediate appearance or a specific detail. The correct sense may be indicated by those background elements.
---
After all your reasoning is finished, provide your final decision as the format 'A: [sense number]', where [sense number] is the index of your chosen sense from the provided list.
"""
```

```python
import pandas as pd
import json

wsd_test_df = pd.read_csv("/workspace/data/set_process/wsd_set_test_labeled.csv")

prompt_list = list()
for index, row in wsd_test_df.iterrows():
    sense_dict = json.loads(row["senses"])
    sense_list = sense_dict.get(row["gold_pos"], [])
    prompt = prompt_template.format(
        word=row["word"],
        context=row["word_phrase"],
        sense_list="\n".join([f" {idx + 1}. {sense}" for idx, sense in enumerate(sense_list)]),
    )
    prompt_list.append(prompt)

wsd_test_df["prompt"] = prompt_list
wsd_test_df.to_csv("/workspace/data/set_process/wsd_set_test_sense_prompt.csv", index=False)
```

#### sub task 1: 상위 문서 k개의 요약문 얻어오기

```python
import trafilatura

def extract_clean_text(url):
    downloaded = trafilatura.fetch_url(url)
    if downloaded:
        # 본문만 추출 (광고, 네비게이션 자동 제거)
        result = trafilatura.extract(downloaded, include_comments=False, include_tables=True)
        return result
    return ""
```

```python
prompt_template="""Ambiguous word: {word}
Entites in Image: 
{entities}
---
You are a linguistic expert. Given a specific target word and 'Entities in Image', your task is to extract and summarize any relevant and helpful information from the given 'Searched Web Content' that can help explain the image. Do not try to describe the image itself if there is no relevant information in the context that can be applied to the image or the 'Entities in Image'. Do not mention the information that is not relevant to the content of the image.
---
Searched Web Content:
- Title: {web_title}
- {text_context}
---
First, Refer to the 'Entities in Image' section to understand the content of the image. Then, read the 'Searched Web Content' section carefully and judge whether it is related to the 'Entities in Image'. Ignore any information that is not relevant to the 'Entities in Image'.
Provide your final answer as a paragraph. The Paragraph should be concise and contain only the essential information.
"""
```

```python
import json
import pandas as pd
from tqdm.auto import tqdm
# 기존에 retrive하여 저장해놓은 목록에서 test 셋에 해당하는 100개 내역 얻어오기
wsd_test_df = pd.read_csv("/workspace/data/set_process/wsd_set_test_labeled.csv")

# k =3
# 요약은 한 번의 LLM 호출로 하나의 인터넷 문서만 처리하도록 구조 적용
retrieval_df = pd.read_csv("/workspace/data/set_process/wsd_set_entire_google_vision_result.csv")

df_dict = {
    "word_index": [],
    "word": [],
    "word_phrase": [],
    "senses": [],
    "gold_image": [],
    "gold_pos": [],
    "gold_sense": [],
    "prompt": []
}
for index, row in tqdm(wsd_test_df.iterrows(), total=len(wsd_test_df)):
    word_index = row["word_index"]
    retrieved_rows = retrieval_df[retrieval_df["word_index"] == word_index]
    retrieved_urls = json.loads(retrieved_rows["web_urls"].iloc[0])
    retrieved_entities = json.loads(retrieved_rows["entities"].iloc[0])
    if len(retrieved_entities) > 0:
        valid_entites = [retrieved_entities[0]["description"]]
        for entity in retrieved_entities[1:]:
            if entity["score"] >= 1.0:
                valid_entites.append(entity["description"])
        valid_entites = "\n".join([f"- {entity}" for entity in valid_entites])
    else:
        valid_entites = ""
    
    retrieved_urls_num = 0
    url_idx = 0
    none_page_titles = list()
    while url_idx < len(retrieved_urls) and retrieved_urls_num < 3:
        cur_url = retrieved_urls[url_idx]["url"]
        cur_title = retrieved_urls[url_idx]["title"]
        if "www.youtube.com" in cur_url:
            web_content = None
        else:
            web_content = extract_clean_text(cur_url)
        
        if web_content is not None:
            web_content_split = web_content.split('\n')
                
            web_content_list = list()
            cur_content = ""
            content_threshold = 4096
            for line in web_content_split:
                if line.strip() != "":
                    cur_content += line.strip() + "\n"
                if len(cur_content) > content_threshold:
                    web_content_list.append(cur_content)
                    cur_content = ""
                
            if len(cur_content) > 0:
                web_content_list.append(cur_content)
                
            for content in web_content_list:
                df_dict["word_index"].append(word_index)
                df_dict["word"].append(row["word"])
                df_dict["word_phrase"].append(row["word_phrase"])
                df_dict["senses"].append(row["senses"])
                df_dict["gold_image"].append(row["gold_image"])
                df_dict["gold_pos"].append(row["gold_pos"])
                df_dict["gold_sense"].append(row["gold_sense"])
                df_dict["prompt"].append(prompt_template.format(word=row["word"], entities=valid_entites, web_title=cur_title, text_context=content))
            retrieved_urls_num += 1
        else:
            none_page_titles.append(cur_title)
        url_idx += 1
    
    # 마지막으로 3개 추가하는 동안 등장한 None 페이지 제목들을 한꺼번에 요약하도록 프롬프트 생성
    if len(none_page_titles) > 0:
        none_page_titles_str = "\n".join([f"- {title}" for title in none_page_titles])
        df_dict["word_index"].append(word_index)
        df_dict["word"].append(row["word"])
        df_dict["word_phrase"].append(row["word_phrase"])
        df_dict["senses"].append(row["senses"])
        df_dict["gold_image"].append(row["gold_image"])
        df_dict["gold_pos"].append(row["gold_pos"])
        df_dict["gold_sense"].append(row["gold_sense"])
        df_dict["prompt"].append(prompt_template.format(word=row["word"], entities=valid_entites, web_title="etc.", text_context=none_page_titles_str))
    
inference_df = pd.DataFrame(df_dict)
inference_df.to_csv("/workspace/data/set_process/wsd_test_summarize_prompt3.csv", index=False)
```

```python
def data_sample_to_example(sample_row, answer_label):
    import json
    
    example_template = """
<start_of_image><image_soft_token><end_of_image>
Word: {word}
---
Noun Senses:
{noun_sense_list}
---
Verb Senses:
{verb_sense_list}
---
Answer: {answer}"""
    word = sample_row["word"]
    sense_dict = json.loads(sample_row["senses"])
    if "noun_senses" in sense_dict:
        noun_sense_list = sense_dict["noun_senses"]
    else:
        noun_sense_list = []
    
    if "verb_senses" in sense_dict:
        verb_sense_list = sense_dict["verb_senses"]
    else:
        verb_sense_list = []
        
    noun_sense_str = "\n".join([f"{i + 1}) {sense}" for i, sense in enumerate(noun_sense_list)])
    verb_sense_str = "\n".join([f"{i + 1}) {sense}" for i, sense in enumerate(verb_sense_list)])
    
    example_prompt = example_template.format(
        word=word,
        noun_sense_list=noun_sense_str,
        verb_sense_list=verb_sense_str,
        answer=answer_label
    )
    
    return example_prompt
```

```python
prompt_template="""
Ambiguous word: {word}
---
{sense_list}

You are an expert linguistic annotator. Image, a specific target word found within the context of that image, and a list of potential senses for that word are given. Each sense is categorized by a Part of Speech (POS): {pos_list}.

Your goal is to determine the correct POS by selecting the sense that most directly aligns with the visual context.
---
Decision Logic & Rules
Follow these rules in order of priority to make your decision:

{rule_list}

Always consider which sense can be most directly and concretely represented in the image. Avoid overcomplicating the reasoning process; focus on the most straightforward interpretation of the visual information in relation to the senses provided.

---
After all your reasoning is finished, provide your final decision as the format 'A: [POS]', where [POS] is your chosen Part of Speech ({pos_list}).
"""

noun_template = """Noun Senses:
{noun_senses}
"""
verb_template = """Verb Senses:
{verb_senses}
"""
adj_template = """Adjective Senses:
{adj_senses}
"""
adv_template = """Adverb Senses:
{adv_senses}
"""

noun_rules = "Every photo features an object. Therefore, consider nouns first."
verb_rules = """Verb is considered when:
* no noun sense is provided.
* the photo depicts an action—either directly or indirectly—rather than the object itself, and if there is a verb sense in the provided list that corresponds to the action shown in the photo."""
adj_rules = """Adjective is considered when:
* the photo can not be considered to depict an object or action of given word.
* the photo depicts a state rather than an action or an object itself, and there is a corresponding sense in the list of adjectives"""
adv_rules = """Adverb is considered when:
* the photo can not be considered to depict an object or action or state of given word."""
```

```python
import pandas as pd
import json

wsd_test_df = pd.read_csv("/workspace/data/test_set_process/wsd_set_100_labeled.csv")

prompt_list = list()
for index, row in wsd_test_df.iterrows():
    sense_dict = json.loads(row["senses"])
    noun_senses = sense_dict.get("n", [])
    verb_senses = sense_dict.get("v", [])
    adj_senses = sense_dict.get("a", [])
    adv_senses = sense_dict.get("r", [])
    
    pos_list = list()
    sense_string = ""
    rule_string = ""
    if len(noun_senses) > 0:
        sense_string += (noun_template.format(noun_senses="\n".join([f"{i + 1}) {sense}" for i, sense in enumerate(noun_senses)])) + "\n---\n")
        #rule_string += noun_rules + '\n\n'
        pos_list.append("noun")
    if len(verb_senses) > 0:
        sense_string += (verb_template.format(verb_senses="\n".join([f"{i + 1}) {sense}" for i, sense in enumerate(verb_senses)])) + "\n---\n")
        rule_string += verb_rules + '\n\n'
        pos_list.append("verb")
    if len(adj_senses) > 0:
        sense_string += (adj_template.format(adj_senses="\n".join([f"{i + 1}) {sense}" for i, sense in enumerate(adj_senses)])) + "\n---\n")
        rule_string += adj_rules + '\n\n'
        pos_list.append("adjective")
    if len(adv_senses) > 0:
        sense_string += (adv_template.format(adv_senses="\n".join([f"{i + 1}) {sense}" for i, sense in enumerate(adv_senses)])) + "\n---\n")
        rule_string += adv_rules + '\n\n'
        pos_list.append("adverb")
    
    prompt = prompt_template.format(
        word=row["word"],
        sense_list=sense_string,
        pos_list=", ".join(pos_list),
        rule_list=rule_string,
    )
    prompt_list.append(prompt)

wsd_test_df["prompt"] = prompt_list
wsd_test_df.to_csv("/workspace/data/test_set_process/wsd_set_100_pos3_prompt.csv", index=False)
```

#### sub task 1: 각 페이지 내용 요약

```python
import trafilatura
import requests

def extract_clean_text(url):
    try:
        response = requests.get(url)
    except:
        return None

    if response.status_code == 200:
        downloaded = trafilatura.fetch_url(url)
        # 본문만 추출 (광고, 네비게이션 자동 제거)
        result = trafilatura.extract(downloaded, include_comments=False, include_tables=True)
        return result
    return None
```

```python
prompt_template="""Ambiguous word: {word}
Context Phrase: {context}
Entites in Image: 
{entities}
---
You are a linguistic expert. Given 'Ambiguous Word', 'Context Phrase', and 'Entities in Image', your task is to extract and summarize any additional and helpful information from the given 'Searched Web Content' that can help explain the context of the image in relation to the 'Ambiguous Word'. Do not try to describe the image itself if there is no relevant information in the 'Searched Web Content' that can be helpful for understanding the 'Ambiguous Word'.
---
Searched Web Content:
- Title: {web_title}
- {text_context}
---
First, Refer to the 'Entities in Image' section to understand the content of the image. Then, read the 'Searched Web Content' section carefully and judge whether it is related to the '{context}', '{word}'. If there is some releavant information for '{word}', generate a word 'Relevant', otherwise generate a word 'Not Relevant'. If the first line is 'Relevant', generate a summary of the helpful information in the 'Searched Web Content' that can explain the context of the image in relation to the 'Ambiguous Word'. If the first line is 'Not Relevant', do not generate any summary and end your generation.
"""
```

```python
import json
import pandas as pd
from tqdm.auto import tqdm
# 기존에 retrive하여 저장해놓은 목록에서 test 셋에 해당하는 100개 내역 얻어오기
wsd_test_df = pd.read_csv("/workspace/data/test_set_process/wsd_set_100_labeled.csv")

# k =3
# 요약은 한 번의 LLM 호출로 하나의 인터넷 문서만 처리하도록 구조 적용
retrieval_df = pd.read_csv("/workspace/data/test_set_process/wsd_set_entire_google_vision_result.csv")

df_dict = {
    "word_index": [],
    "word": [],
    "word_phrase": [],
    "senses": [],
    "gold_image": [],
    "gold_pos": [],
    "gold_sense": [],
    "prompt": []
}
for index, row in tqdm(wsd_test_df.iterrows(), total=len(wsd_test_df)):
    word_index = row["word_index"]
    retrieved_rows = retrieval_df[retrieval_df["word_index"] == word_index]
    retrieved_urls = json.loads(retrieved_rows["web_urls"].iloc[0])
    retrieved_entities = json.loads(retrieved_rows["entities"].iloc[0])
    if len(retrieved_entities) > 0:
        valid_entites = [retrieved_entities[0]["description"]]
        for entity in retrieved_entities[1:]:
            if entity["score"] >= 1.0:
                valid_entites.append(entity["description"])
        valid_entites = "\n".join([f"- {entity}" for entity in valid_entites])
    else:
        valid_entites = ""
    
    retrieved_urls_num = 0
    url_idx = 0
    none_page_titles = list()
    while url_idx < len(retrieved_urls) and retrieved_urls_num < 3:
        cur_url = retrieved_urls[url_idx]["url"]
        cur_title = retrieved_urls[url_idx]["title"]
        if "www.youtube.com" in cur_url:
            web_content = None
        else:
            web_content = extract_clean_text(cur_url)
        
        if web_content is not None:
            web_content_split = web_content.split('\n')
                
            web_content_list = list()
            cur_content = ""
            content_threshold = 4096
            for line in web_content_split:
                # 현재 line이 threshold를 넘는지 확인
                if len(line.strip()) < content_threshold:
                    next_content = cur_content + "\n" + line.strip()
                    if len(next_content) > content_threshold:
                        web_content_list.append(cur_content)
                        cur_content = line.strip()
                    else:
                        cur_content = next_content

            if len(cur_content) > 0:
                web_content_list.append(cur_content)
                
            for content in web_content_list:
                df_dict["word_index"].append(word_index)
                df_dict["word"].append(row["word"])
                df_dict["word_phrase"].append(row["word_phrase"])
                df_dict["senses"].append(row["senses"])
                df_dict["gold_image"].append(row["gold_image"])
                df_dict["gold_pos"].append(row["gold_pos"])
                df_dict["gold_sense"].append(row["gold_sense"])
                df_dict["prompt"].append(prompt_template.format(word=row["word"], entities=valid_entites, context=row["word_phrase"], web_title=cur_title, text_context=content))
            retrieved_urls_num += 1
        else:
            none_page_titles.append(cur_title)
        url_idx += 1
    
    # 마지막으로 3개 추가하는 동안 등장한 None 페이지 제목들을 한꺼번에 요약하도록 프롬프트 생성
    if len(none_page_titles) > 0:
        none_page_titles_str = "\n".join([f"- {title}" for title in none_page_titles])
        df_dict["word_index"].append(word_index)
        df_dict["word"].append(row["word"])
        df_dict["word_phrase"].append(row["word_phrase"])
        df_dict["senses"].append(row["senses"])
        df_dict["gold_image"].append(row["gold_image"])
        df_dict["gold_pos"].append(row["gold_pos"])
        df_dict["gold_sense"].append(row["gold_sense"])
        df_dict["prompt"].append(prompt_template.format(word=row["word"], entities=valid_entites, context=row["word_phrase"], web_title="etc.", text_context=none_page_titles_str))
    
inference_df = pd.DataFrame(df_dict)
inference_df.to_csv("/workspace/data/test_set_process/wsd_100_summarize_prompt3.csv", index=False)
```

#### sub task 2: 요약한 정보로 sense labeling하기

```python
prompt_template="""Text Context: {context}
Ambiguous Word: {word}
---
Sense List:
{sense_list}
---
Image Web Search Results:
{web_search_results}
---
You are an expert linguistic annotator. A specific target word found within the context of a given text, and a list of potential senses for that word are provided. Your task is to determine the correct sense by selecting the one that most directly aligns with the context and its background event.
---
Decision Logic & Rules
Follow these rules in order of priority to make your decision:

First, Use web search results to determine whether there is any background information associated with the given image that cannot be discerned solely from its visual elements. If such information exists, prioritize the sense that best fits this background information, even if it is not the most direct match for the visual content.

Directness of Meaning: If no additional background information can be obtained from the web search results, choose the sense that provides the most direct and specific fit for the visual and linguistic context. Even if a perfect match does not exist, select the sense that is the closest indirect match.
---
After all your reasoning is finished, provide your final decision as the format 'A: [sense number]' and end your generation. In here, [sense number] is the index of your chosen sense from the provided list.
"""
```

```python
import pandas as pd
import json

wsd_test_df = pd.read_csv("/workspace/data/test_set_process/wsd_set_100_labeled.csv")
web_search_df = pd.read_csv("/workspace/data/test_set_process/inference/wsd_100_summarize2_gemma3-27b.csv")

prompt_list = list()
for index, row in wsd_test_df.iterrows():
    sense_dict = json.loads(row["senses"])
    sense_list = sense_dict.get(row["gold_pos"], [])
    web_search_results = web_search_df.loc[web_search_df["word"] == row["word"], "generated_text"].tolist()
    
    cropped_web_search_results = ""
    crop_threshold = 3000
    while len(cropped_web_search_results) < crop_threshold and len(web_search_results) > 0:
        next_result = web_search_results.pop(0).strip()
        if next_result != "" and "Not Relevant" not in next_result:
            cropped_web_search_results += (next_result + "\n")
    
    prompt = prompt_template.format(
        word=row["word"],
        context=row["word_phrase"],
        sense_list="\n".join([f" {idx + 1}. {sense}" for idx, sense in enumerate(sense_list)]),
        web_search_results=cropped_web_search_results
    )
    prompt_list.append(prompt)

wsd_test_df["prompt"] = prompt_list
wsd_test_df.to_csv("/workspace/data/test_set_process/wsd_set_100_sense_search2_prompt.csv", index=False)
```

## 모호한 문장 생성하기

```python
prompt_template="""the {pos} '{word}' that can be interpreted as:
{sense_list}
---
Generate a natural ambiguous sentence using polysemous word '{word}', once so that the sentence can be interpreted in more than one meaning.
In the generated sentence, the {pos} '{word}' should be interpreted as '{gold_sense}'. However, the sentence should not imply any of the given definitions too strongly.
Suggest your final sentence in the format 'Generated Sentence: [your sentence]'. Do not include any explanation, and end your generation after providing the sentence."""
```

```python
import pandas as pd
import json

wsd_test_df = pd.read_csv("/workspace/data/test_set_process/wsd_set_100_labeled.csv")
pos_mapping = {
    "n": "noun",
    "v": "verb",
    "a": "adjective",
    "r": "adverb"
}

prompt_list = list()
for index, row in wsd_test_df.iterrows():
    sense_dict = json.loads(row["senses"])
    sense_list = sense_dict.get(row["gold_pos"], [])
    
    prompt = prompt_template.format(
        word=row["word"],
        pos=pos_mapping.get(row["gold_pos"], row["gold_pos"]),
        sense_list="\n".join([f" {idx + 1}. {sense}" for idx, sense in enumerate(sense_list)]),
        gold_sense=sense_list[row["gold_sense"] - 1]
    )
    prompt_list.append(prompt)

wsd_test_df["prompt"] = prompt_list
wsd_test_df.to_csv("/workspace/data/test_set_process/wsd_set_100_ambiguous_sentence_prompt.csv", index=False)
```

## pos 판별 task

```python
prompt_template="""
Ambiguous word: {word}
---
{sense_list}

You are an expert linguistic annotator. Image, a specific target word found within the context of that image, and a list of potential senses for that word are given. Each sense is categorized by a Part of Speech (POS): {pos_list}.

Your goal is to determine the correct POS by selecting the sense that most directly aligns with the visual context.
---
Decision Logic & Rules
Follow these rules in order of priority to make your decision:

{rule_list}

Always consider which sense can be most directly and concretely represented in the image. Avoid overcomplicating the reasoning process; focus on the most straightforward interpretation of the visual information in relation to the senses provided.

---
After all your reasoning is finished, provide your final decision as the format 'A: [POS]', where [POS] is your chosen Part of Speech ({pos_list}).
"""

noun_template = """Noun Senses:
{noun_senses}
"""
verb_template = """Verb Senses:
{verb_senses}
"""
adj_template = """Adjective Senses:
{adj_senses}
"""
adv_template = """Adverb Senses:
{adv_senses}
"""

noun_rules = "Every photo features an object. Therefore, consider nouns first."
verb_rules = """Verb is considered when:
* no noun sense is provided.
* the photo depicts an action—either directly or indirectly—rather than the object itself, and if there is a verb sense in the provided list that corresponds to the action shown in the photo."""
adj_rules = """Adjective is considered when:
* the photo can not be considered to depict an object or action of given word.
* the photo depicts a state rather than an action or an object itself, and there is a corresponding sense in the list of adjectives"""
adv_rules = """Adverb is considered when:
* the photo can not be considered to depict an object or action or state of given word."""
```

```python
import pandas as pd
import json

wsd_test_df = pd.read_csv("/workspace/data/test_set_process/wsd_set_entire.csv")

prompt_list = list()
for index, row in wsd_test_df.iterrows():
    sense_dict = json.loads(row["senses"])
    noun_senses = sense_dict.get("n", [])
    verb_senses = sense_dict.get("v", [])
    adj_senses = sense_dict.get("a", [])
    adv_senses = sense_dict.get("r", [])
    
    pos_list = list()
    sense_string = ""
    rule_string = ""
    if len(noun_senses) > 0:
        sense_string += (noun_template.format(noun_senses="\n".join([f"{i + 1}) {sense}" for i, sense in enumerate(noun_senses)])) + "\n---\n")
        #rule_string += noun_rules + '\n\n'
        pos_list.append("noun")
    if len(verb_senses) > 0:
        sense_string += (verb_template.format(verb_senses="\n".join([f"{i + 1}) {sense}" for i, sense in enumerate(verb_senses)])) + "\n---\n")
        rule_string += verb_rules + '\n\n'
        pos_list.append("verb")
    if len(adj_senses) > 0:
        sense_string += (adj_template.format(adj_senses="\n".join([f"{i + 1}) {sense}" for i, sense in enumerate(adj_senses)])) + "\n---\n")
        rule_string += adj_rules + '\n\n'
        pos_list.append("adjective")
    if len(adv_senses) > 0:
        sense_string += (adv_template.format(adv_senses="\n".join([f"{i + 1}) {sense}" for i, sense in enumerate(adv_senses)])) + "\n---\n")
        rule_string += adv_rules + '\n\n'
        pos_list.append("adverb")
    
    prompt = prompt_template.format(
        word=row["word"],
        sense_list=sense_string,
        pos_list=", ".join(pos_list),
        rule_list=rule_string,
    )
    prompt_list.append(prompt)

wsd_test_df["prompt"] = prompt_list
wsd_test_df.to_csv("/workspace/data/test_set_process/wsd_set_entire_pos.csv", index=False)
```
